```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv("netflix_clean.csv")
df.head()
```

```python
release_years=df['release_year'].to_numpy()
```

```python
# mean
avg_year=np.mean(release_years)
print(f'average release year : {avg_year}')
```

```python
# median
median_year=np.median(release_years)
print(f'average release year : {median_year}')
```

```python
std_year=np.std(release_years)
print(f'std year: {std_year}')
```

```python
print("Oldest Release Year :", np.min(release_years))
print("Newest Release Year :", np.max(release_years))
```

```python
release_range = np.ptp(release_years)

print("Range:", release_range)
```

```python
recent_content = np.sum(release_years >= 2015)

print("Content Released After 2015:", recent_content)
```
