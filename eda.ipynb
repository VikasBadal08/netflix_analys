```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv("netflix_clean.csv")
df.head()
```

```python
print(f"total  no of movies :{df['type'].value_counts()['Movie']}")
print(f"total  no of TV Shows :{df['type'].value_counts()['TV Show']}")
type_count=df['type'].value_counts()
plt.pie(type_count,labels=type_count.index,
        colors=['lightblue','blue'],
        autopct="%1.1f%%")
plt.legend()
plt.title("Movies vs Tv show Distributions")
plt.show()
```

```python
# top 10 countries producing netflix content 
top_country=df['country'].value_counts().head(10)
print(f"Top 10 Countries Producing Netflix Content:{top_country}")

plt.figure(figsize=(12,6))
plt.bar(top_country.index,top_country.values)
plt.xlabel('country')
plt.title("Top 10 Countries Producing Netflix Content")
plt.ylabel("No of content")
plt.tight_layout()
plt.show()

# which country dominates netflix 
print(F'dominating country content wise is {top_country.idxmax()}')
```

```python
# Content Added Every Year
year_wise_count=df['year_added'].value_counts()
print(f"year wise content:{year_wise_count}")

plt.figure(figsize=(12,8))
plt.plot(year_wise_count.index,year_wise_count.values,marker='o')
plt.xlabel("year")
plt.ylabel("no of content")
plt.title("year wise content")
plt.xticks(year_wise_count.index)
plt.grid()
plt.show()
```
