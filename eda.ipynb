```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv("netflix_clean.csv")
df.head()
```

```python
print(f"total  no of movies :{df['type'].value_counts()['Movie']}")
print(f"total  no of TV Shows :{df['type'].value_counts()['TV Show']}")
type_count=df['type'].value_counts()
plt.pie(type_count,labels=type_count.index,
        colors=['lightblue','blue'],
        autopct="%1.1f%%")
plt.legend()
plt.title("Movies vs Tv show Distributions")
plt.show()
```

```python
# top 10 countries producing netflix content 
top_country=df['country'].value_counts().head(10)
print(f"Top 10 Countries Producing Netflix Content:{top_country}")

plt.figure(figsize=(12,6))
plt.bar(top_country.index,top_country.values)
plt.xlabel('country')
plt.title("Top 10 Countries Producing Netflix Content")
plt.ylabel("No of content")
plt.tight_layout()
plt.show()

# which country dominates netflix 
print(F'dominating country content wise is {top_country.idxmax()}')
```

```python
# Content Added Every Year
year_wise_count=df['year_added'].value_counts()
print(f"year wise content:{year_wise_count}")

plt.figure(figsize=(12,8))
plt.plot(year_wise_count.index,year_wise_count.values,marker='o')
plt.xlabel("year")
plt.ylabel("no of content")
plt.title("year wise content")
plt.xticks(year_wise_count.index)
plt.grid()
plt.show()
```

```python
# extract genres
genre_df=df.copy()
genre_df['listed_in']=genre_df['listed_in'].str.split(",")
```

```python
genre_df=genre_df.explode('listed_in')
# explode()= transfer each element of a list like structure into its own seprate row 
# example comedy , drame count one without explode 
# but with explode it counts in separately 
```

```python
genre_df['listed_in']=genre_df['listed_in'].str.strip()
top_genres=genre_df['listed_in'].value_counts().head(20)
top_genres
```

```python
# bar chart
plt.Figure(figsize=(12,8))
plt.barh(
    top_genres.index,
    top_genres.values,
    color='skyblue',
    edgecolor='black'
)
plt.title("top  20 genres on netflix ")
plt.xlabel("no of titles")
plt.ylabel('genres')
plt.tight_layout()
plt.show()
```

```python
genre_year=genre_df.groupby(['release_year','listed_in']).size()
genre_year
```

```python
genre_year=genre_year.unstack(fill_value=0)
genre_year
```

```python
top5=genre_df['listed_in'].value_counts().head(5).index
top5
```

```python
genre_year=genre_year[genre_year.index>=2015]
plt.figure(figsize=(15,8))
for genre in top5:
    plt.plot(
        genre_year.index,
        genre_year[genre],
        marker='*',
        label=genre
    )
plt.title("genres by time")
plt.xlabel("year")
plt.ylabel("no of titles")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
country_df=df.copy()
country_df=country_df[country_df['country']!="not Available"]
country_df['country']=country_df['country'].str.split(",")
country_df=country_df.explode('country')
country_df['country']=country_df['country'].str.strip()
```

```python
top_10_cur=country_df['country'].value_counts().head(10).index
top_10_cur
```

```python
country_df=country_df[country_df['country'].isin(top_10_cur)]
avg_year=(country_df.groupby('country')['release_year'].mean().sort_values(ascending=False))
print(avg_year)
```

```python
plt.Figure(figsize=(12,6))
plt.bar(
    avg_year.index,
    avg_year.values,
    color='red',
    edgecolor='blue'
)
plt.title("Top 10 Countries by Average Release Year")
plt.xlabel("Country")
plt.ylabel("Average Release Year")
plt.ylim(2010, 2021)
plt.xticks(rotation=45, ha="right")

plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()

print(f"south korea have newer content")
```
