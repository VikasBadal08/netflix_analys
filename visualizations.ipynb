```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv("netflix_clean.csv")
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x="type")

plt.title("Movies vs TV Shows")
plt.xlabel("Content Type")
plt.ylabel("Count")
plt.show()
```

```python
movie_df = df[df["type"]=="Movie"].copy()

movie_df["duration"] = (
    movie_df["duration"]
    .str.replace(" min","")
    .astype(int)
)

sns.histplot(movie_df["duration"], bins=20)

plt.title("Movie Duration Distribution")
plt.xlabel("Duration (Minutes)")
plt.show()
```

```python
top_country = df["country"].value_counts().head(10).index

country_df = df[df["country"].isin(top_country)]

table = pd.crosstab(country_df["country"], country_df["type"])

sns.heatmap(table, annot=True, cmap="Blues")

plt.title("Country vs Content Type")
plt.show()
```

```python
top_country = df["country"].value_counts().head(10)

sns.barplot(x=top_country.values, y=top_country.index)

plt.title("Top 10 Countries")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```
